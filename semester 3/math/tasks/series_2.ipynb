```python
data_geysir = pd.read_csv("./geysir.dat", delimiter=' ', index_col=0)
data_geysir
```

```python

# (2 2 1)
#  â”‚ â”‚ â””â”€â”€ select first figure
#  â”‚ â””â”€â”€â”€â”€ two columns
#  â””â”€â”€â”€â”€â”€â”€ two rows
plt.subplot(221)
data_geysir["Zeitspanne"].plot(kind="hist", edgecolor="black")
plt.xlabel("10 Klassen")
plt.subplot(222)
data_geysir["Zeitspanne"].plot(kind="hist",
                               bins=20,
                               edgecolor="black")
plt.xlabel("20 Klassen")
plt.subplot(223)
# select specific points
data_geysir["Zeitspanne"].plot(kind="hist",
                               bins=np.arange(41, 107, 11),
                               edgecolor="black")
plt.xlabel("Klassengrenzen 41, 52 , 63, 74 , 85, 96")
plt.tight_layout()

print("a)")
plt.show()
```

Anhand den ersten zwei Histogrammen kann erkannt werden, dass nicht nur bei ca. 55 Minuten hÃ¤ufig die Intervallen sind, sondern auch zwischen 75 - 85 Minuten entstehen.

Bei Figur 3 wurde die KlassengrÃ¶sse falsch gewÃ¤hlt und somit sind Informationen verdeckt, welche bei kleineren Klassen einfacher zu erkennen sind.

```python
# b)
plt.subplot(221)
data_geysir["Eruptionsdauer"].plot(kind="hist",
bins = 5,
edgecolor="black")
plt.xlabel("5 Klassen")
plt.subplot(222)
data_geysir["Eruptionsdauer"].plot(kind="hist",
                               bins=10,
                               edgecolor="black")
plt.xlabel("10 Klassen")
plt.subplot(223)
data_geysir["Eruptionsdauer"].plot(kind="hist",
                               bins=20,
                               edgecolor="black")
plt.xlabel("20 Klassen")
plt.subplot(224)
data_geysir["Eruptionsdauer"].plot(kind="hist",
                              bins=30,
                               edgecolor="black")
plt.xlabel("30 Klassen")
plt.tight_layout()

print("a)")
plt.show()
```

Entweder dauert die Eruption $<1$ Minuten oder verteilt sich im Bereich $3$ bis $4.6$ Minuten. Bei den gewÃ¤hlten KlassengrÃ¶ssen konnte bei jedem die auffÃ¤lligsten Informationen gefunden werden.

## Aufgabe 2.3

```python
grades = pd.Series([4.2, 2.3, 5.6, 4.5, 4.8, 3.9, 5.9, 2.4, 5.9, 6, 4, 3.7, 5, 5.2, 4.5, 3.6, 5, 6, 2.8, 3.3, 5.5, 4.2, 4.9, 5.1])

# a)

grades_edited = grades.copy()
grades_edited[1] = 1.1
grades_edited[6] = 6
grades_edited[18] = 3.0

print(f"Original -> median: {grades.median()} mean: {grades.mean()}")
print(f"Edited -> median: {grades_edited.median()} mean: {grades_edited.mean()}")
```

```python
# b)

plt.subplot(221)
grades.plot(kind="hist")
plt.ylabel("Anzahl")

plt.subplot(222)
grades.plot(kind="box")

plt.subplot(223)
grades_edited.plot(kind="hist")
plt.ylabel("Anzahl")

plt.subplot(224)
grades_edited.plot(kind="box")
```
