# **Wahrscheinlichkeitsmodell**

## Bedingte Wahrscheinlichkeit

## Binomial- und Poissonverteilung

### Binominalverteilung

$$
P(X=k)=\left(\begin{matrix}n \\ k \end{matrix}\right)\cdot p^k\cdot (1-p)^{n-k}
$$

$$
P(15 \leq X \leq 25)=\sum_{k=15}^{25}P(X=k)
$$

```python
import scipy.stats

k = 20
n = 100
p = 0.2

scipy.stats.binom.pmf(k,n,p)
```

```python
n, p = 100, 0.2

ran = np.arange(n+1)
data = scipy.stats.binom.pmf(ran,n,p)


plt.stem(ran[15:26],data[15:26]) # P(15 <= X <= 25)
plt.stem(ran,data)
```
