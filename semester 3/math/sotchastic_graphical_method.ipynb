# Stochastik - Graphische Methode

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

methodeA = pd.Series([79.98, 80.04, 80.02, 80.04, 80.03, 80.03, 80.04, 79.97, 80.05, 80.03, 80.02, 80.00, 80.02])
methodeB = pd.Series([80.02, 79.94, 79.98, 79.97, 79.97, 80.03, 79.95, 79.97])
```

## Histogramm

```python
methodeA.plot(kind="hist", edgecolor="black")
plt.title("Histogramm von Methode A")
plt.xlabel("methodeA")
plt.ylabel("Haeufigkeit")
plt.show()
```

### Kumulatives Histogramm

```python
methodeA.plot(kind="hist", cumulative=True, histtype="step", density=True, bins = 100, edgecolor="black")
plt.show()
```

## Boxplot

```python
methodeA.plot(kind="box", title="Boxplot Methode A")
plt.show()
```

### Boxplot mit zwei Werten

```python
methode = pd.DataFrame({
  "methodeA": methodeA,
  "methodeB": methodeB
})

methode.plot(kind="box", title="Boxplot von Methode A und B")

plt.show()
```
